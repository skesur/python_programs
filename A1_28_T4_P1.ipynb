```python
# multilevel inheritance

class Father:
    def __init__(self,a):
        print("Inside Father Class")
        self.a=a
    def display(self):
        print("Father Instance Method")
        print(self.a)
class Child(Father):
    def __init__(self,b,a):
        super().__init__(a)
        print("Inside Child Class")
        self.b=b
    def show(self):
        print("Child Instance Method")
class GrandChild(Child):
    def __init__(self,b,a):
        super().__init__(b,a)
        print("Inside GrandChild Class")
    def show(self):
        print("GrandChild Instance Method")
    
g = GrandChild(3,6)
g.display()
g.show()
print(1)
```

```python
# hierarchical inheritance

class Father:
    def __init__(self):
        print("Inside Father Class")
        super().__init__()
    def display(self):
        print("Father Instance Method")
class Son(Father):
    def __init__(self):
        print("Inside Son Class")
        super().__init__()
    def show(self):
        print("Son Instance Method")           
class Daughter(Father):
    def __init__(self):
        print("Inside Daughter Class")
        super().__init__()
    def show(self):
        print("Daughter Instance Method")  
s=Son()
d=Daughter()
s.display()
d.display()
```

```python
# operator overloading

class A:
    def __init__(self,a):
        self.a=a
    def __str__(self):
        return "Operator Overloading"
x = A(5)    
print(x) 

class A:
    def __init__(self,a):
        self.a=a
    def display(self):
        print(self.a)
    def __add__(self,other):
        return self.a+other.a
    def __sub__(self,other):
        return self.a-other.a
    def __mul__(self,other):
        return self.a*other.a
    def __truediv__(self,other):
        return self.a/other.a
a1=A(4)
a2=A(2)
print(a1+a2)
print(a1-a2)
print(a1*a2)
print(a1/a2)


# a1==a2 - __eq__
# a1!=a2 - __ne__
# a1>a2 - __gt__
# a1<a2 - __lt__
# a1>=a2 - __ge__
# a1<=a2 - __le__
```
