```python
# write a program which takes three different int from user and print those numbers in assending 
# and descending order

a1 = int(input("Enter first number : "))
a2 = int(input("Enter second number : "))
a3 = int(input("Enter third number : "))

print("\n")
print("Ascending order : ")
if a1<a2 and a1<a3 :
    print(a1)
    if a2<a3 :
        print(a2)
        print(a3)
    else :
        print(a3)
        print(a2)
elif a2<a1 and a2<a3 :
    print(a2)
    if a1<a3 :
        print(a1)
        print(a3)
    else :
        print(a3)
        print(a1)
elif a3<a1 and a3<a1 :
    print(a3)
    if a1<a2 :
        print(a1)
        print(a2)
    else :
        print(a2)
        print(a1)
elif a1==a2==a3 :
    print(a1)
    print(a2)
    print(a3)

print("\n")
print("Descending order : ")
if a1>a2 and a1>a3 :
    print(a1)
    if a2>a3 :
        print(a2)
        print(a3)
    else :
        print(a3)
        print(a2)
elif a2>a1 and a2>a3 :
    print(a2)
    if a1>a3 :
        print(a1)
        print(a3)
    else :
        print(a3)
        print(a1)
elif a3>a1 and a3>a1 :
    print(a3)
    if a1>a2 :
        print(a1)
        print(a2)
    else :
        print(a2)
        print(a1)
elif a1==a2==a3 :
    print(a1)
    print(a2)
    print(a3)        
                
```
