```python
# WAP which will return the sum of numbers from the list except the number 13 and the number that comes immediately after13
# output should be zero for empty list

l = eval(input("Enter the list : "))
sum=0
for i in range(len(l)):
    if i==len(l)-1:
        if l[i]==13:
            sum=sum+0
        else:
            sum=sum+l[i]
    elif i!=len(l)-1 and i!=0:
        if l[i]==13 or l[i-1]==13:
            sum=sum+0
        else:
            sum=sum+l[i]
    elif i==0:
        if l[i]==13:
            sum=sum+0
        else:
            sum=sum+l[i]
print(f"sum: {sum}")        
```
