```python
import numpy as np

class GuessingGame:
    n=0
    a=0
    d=0
    def __init__(self,n,a,d):
        self.n=n
        self.a=a
        self.d=d
        self.start()
    def start(self):
        print("Start Guessing : ")
        while self.a>0:
            i = self.accept()
            self.validate(i)
            if i==self.n:
                print("You win!!!")
                return
            self.a=self.a-1
        print("You lose!!!")  
        print("Number was ",self.n)
    def accept(self):
        return int(input("Enter your guess : "))
    def validate(self,i):
        if i==self.n:
            self.display("Correct guess")
        elif i>self.n:
            self.display("Too high")
        elif i<self.n:
            self.display("Too low")
    def display(self,p):
        print(p)

print("Start Guessing Game : ")
d = int(input("Choose Difficulty : 1=Easy / 2=Hard : "))

if d==1 :
    n = np.random.randint(0,100)
    a = 6
    g = GuessingGame(n,a,d)
elif d==2 :
    n = np.random.randint(0,1000)
    a = 9
    g = GuessingGame(n,a,d)
else :
    print("Input is Invalid")
    
```
