```python
# Write a program that converts Roman numerals into ordinary numbers. Roman
# numerals are represented by the following symbols and their respective values:
#  M = 1000
#  D = 500
#  C = 100
#  L = 50
#  X = 10
#  V = 5
#  I = 1
# Additionally, special cases like IV (4), XL (40) etc. should also be considered. The
# program should take a Roman numeral as input and output its equivalent ordinary
# number.

roman = {"I":1,"IV":4,"V":5,"IX":9,"X":10,"XL":40,"L":50,"XC":90,"C":100,"CD":400,"D":500,"CM":900,"M":1000,}
user = input("Enter a roman number : ")
num = 0
i = 0
while i<len(user):
    if user[i:i+2] in roman:
        num+=roman[user[i:i+2]]
        i+=2
    elif user[i:i+1] in roman:
        num+=roman[user[i:i+1]]
        i+=1
print(num)        
```

```python
n=[1000,900,500,400,100,90,50,40,10,9,5,4,1]
roman=["M","CM","D","CD","C","XC","L","XL","X","IX","V","IV","I"]
user=int(input("Enter a roman number : "))
i=0
result=''

while user>0:
    for j in range(user//n[i]):
        result+=roman[i]
        user-=n[i]
    i+=1
print(result)
```

```python
b = [
"Inferno,Dan Brown,Thriller,450,4.5",
"Atomic Habits,James Clear,SelfHelp,550,4.8",
"Ikigai,Hector Garcia,SelfHelp,300,4.2",
"Dune,Frank Herbert,ScienceFiction,600,4.6",
"1984,George Orwell,ScienceFiction,500,4.7",
"Sapiens,Yuval Noah Harari,History,650,4.9",
"The Hobbit,J.R.R. Tolkien,Fantasy,480,4.8",
"Becoming,Michelle Obama,Biography,700,4.7",
"The Alchemist,Paulo Coelho,Fiction,350,4.3",
"To Kill a Mockingbird,Harper Lee,Fiction,400,4.6"
]

import json

d={}
for i in b:
    n,a,g,p,r = i.split(",")
    d[n]={"author":a,"genre":g,"price":p,"rating":r}
print(d)
print()

# a. Display the average price of all books.
total=0
for i in d.values():
    total += int(i["price"])
print(total/len(d))    
print()

# b. Display the highest-rated book (title, author, rating).
rating=sorted(d.items(),key=lambda x:x[1]["rating"],reverse=True)
print(rating[0][0])
print(rating[0][1]["author"])
print(rating[0][1]["rating"])
print()

# c. Group books by genre and show the count of books in each genre.
genre={}
for i in d.values():
    if i["genre"] not in genre:
        genre[i["genre"]] = 1
    else:
        genre[i["genre"]] +=1
print(genre)        
print()

# d. Display all ScienceFiction books sorted by rating (descending).
scifi={}
for i in d:
    if d[i]["genre"]=="ScienceFiction":
        scifi[i]=d[i]["rating"]
print(sorted(scifi.items(),key=lambda x:x[1],reverse=True))        
print()

# e. Display the most expensive book in each genre.
me={}
for i in d:
    g=d[i]["genre"]
    p=d[i]["price"]
    if g not in me or d[i]["price"] > me[g]["price"]:
        me[g]={"book":i,"price":p}
print(me)
```
