```python
# WAP n children standing in a line each child is asigned a rating value given in the list of int you giving candies to
# this children. Subjected too following requirements 
# 1. Each child must have atleast one candy
# 2. Children with higher rating get more candies then their neighbors
# Return the minimum no. of candies you need to have to distribute among the children

l = eval(input("Enter list of ratings : "))
candy=[1]*len(l)
n=len(l)
for i in range(1,n):
    if l[i]>l[i-1]:
        candy[i]=candy[i-1]+1
for i in range(n-2,-1,-1):
    if l[i]>l[i+1]:
        candy[i]=max(candy[i],candy[i+1]+1)
print(sum(candy))
```
