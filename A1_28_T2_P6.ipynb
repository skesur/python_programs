```python
# WAP to print given number 0 to 1000 in words

d1 = {1:"One",2:"Two",3:"Three",4:"Four",5:"Five",6:"Six",7:"Seven",8:"Eight",9:"Nine",10:"Ten",11:"Eleven",12:"Twelve",
     13:"Thirteen",14:"Fourteen",15:"Fifteen",16:"Sixteen",17:"Seventeen",18:"Eighteen",19:"Nineteen",20:"Twenty",
     30:"Thirty",40:"Forty",50:"Fifty",60:"Sixty",70:"Seventy",80:"Eighty",90:"Ninety",100:"Hundred",1000:"Thousand"}

n = int(input("Enter the number : "))
t = n
s1 = str(n)
s2 = ""
c = len(s1)
for i in range(c,0,-1):
    r = t%(10**(i))
    s = str(r)
    if i==4:
        if s[-1]==0 and s[-2]==0 and s[-3]==0:
            s2=s2+d1[1000]
            break
    if i==3:
        if s[-1]==0 and s[-2]==0:
            s2=s2+" "+d1[int(s[0])]+" "+d1[100]    
    if i==2:
        if s[-1]==0:
            s2=s2+" "+d1[r]
    if i==1:
        s2=s2+" "+d1[r]
print(s2)        
```
