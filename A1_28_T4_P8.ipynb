```python
import numpy as np

sales = np.array([
 [[120,10], [140,12], [130,11], [150,13], [135,10]],
 [[80,15], [90,18], [85,16], [88,17], [92,19]],
 [[200,8], [210,9], [205,10], [215,8], [220,7]],
 [[60,20], [65,22], [70,21], [68,23], [72,24]],
 [[150,12], [155,13], [148,14], [160,15], [158,12]],
 [[230,6], [240,7], [235,6], [245,8], [250,7]]
]) 

class BaseSalesData:
    def __init__(self,sales):
        self.sales=sales
        self.products=self.sales.shape[0]
        self.regions=self.sales.shape[1]
    
class SalesStabilityAnalysis(BaseSalesData):
    def net_units(self):
        net=[]
        for p in range(self.products):
            row=[]
            for r in range(self.regions):
                row.append(self.sales[p][r][0]-self.sales[p][r][1])
            net.append(row)
        return np.array(net)
    def risk_products(self):
        risky=[]
        for p in range(self.products):
            for r in range(self.regions):
                if self.sales[p][r][1]>15:
                    risky.append(p)
                    break
        return risky  
    def regional_reliability(self):
        net=self.net_units()
        avg=[]
        for r in range(self.regions):
            s=0
            for p in range(self.products):
                s+=net[p][r]
            avg.append(s/self.products)
        return avg    
    def most_stable_product(self):
        net=self.net_units()
        min_values=[]
        for p in range(self.products):
            min_values.append(np.min(net[p]))
        return np.argmax(min_values)    
    def normalize_returns(self):
        for p in range(self.products):
            for r in range(self.regions):
                if 14<self.sales[p][r][1]<=16:
                    self.sales[p][r][1]=14
        return self.net_units()            
    def stability_index(self):
        net=self.net_units()
        index=[]
        for p in range(self.products):
            total=0
            for r in range(self.regions):
                total+=net[p][r]
            avg=total/self.regions
            index.append((avg/200)*100)
        return index    
    def forecast_readiness(self):
        index=self.stability_index()
        status=[]
        for i in index:
            if i>=75:
                status.append("High")
            elif i>=60:
                status.append("Moderate")
            elif i>=50:
                status.append("Low")
            else:
                status.append("Unstable")
        return status        
    def planning_priority(self):
        index=self.stability_index()
        temp=list(enumerate(index))
        temp.sort(key=lambda x:x[1],reverse=True)
        order=[]
        for i in temp:
            order.append(i[0])
        return order    
        
sa = SalesStabilityAnalysis(sales)
print("Net Units:",sa.net_units(),"\n")
print("Risk Products: ",sa.risk_products(),"\n")
print("Regional Reliability: ",sa.regional_reliability(),"\n")
print("Most Stable Product ID: ",sa.most_stable_product(),"\n")
print("After Return Normalization:\n", sa.normalize_returns(),"\n")
print("Stability Index:", sa.stability_index(),"\n")
print("Forecast Readiness:", sa.forecast_readiness(),"\n") 
print("Planning Priority:", sa.planning_priority()) 
```
