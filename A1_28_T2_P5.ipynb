```python
#WAP to find longest common prefix from given list of strings, if there no common prefix then return -1
#["less","lesson","lessonplan","left"] -> output="le"

l = sorted(eval(input("Enter the list : ")))
small=min(len(l[0]),len(l[-1]))
pre=""
for i in range(small):
    if l[0][i] == l[-1][i]:
        pre += l[0][i]
    else:
        break;
if pre:
    print(pre)
else:
    print(-1)     
```

```python
# WAP to count the number of special elements in give list
# An element is special if removal of that element makes the list balanced
# The list will be balanced, if sum of even index elements is equal to sum of odd index elements
# The given list : [5,5,2,5,8] -> output should be 5(0),5(1) => 2

l1 = eval(input("Enter the list : "))
c = 0
for i in range(len(l1)):
    l1i = l1[i]
    del l1[i]
    sum1 = 0
    sum2 = 0
    for j in range(len(l1)):
        if j%2==0:
            sum2 +=l1[j]
        if j%2!=0:
            sum1 +=l1[j]
    if sum1==sum2:
        c+=1
    l1.insert(i,l1i)
print(c)
```

```python
# Rail fence cipher 
# msg = "this is python"
# key = 3
# part 1 -> 0,3,6,9,12 
# part 2 -> 1,4,7,10,13
# part 3 -> 2,5,8,11
# output = tssya/h__tn/iiph = tssyah__tniiph

s = input("Enter the string : ")
k = int(input("Enter the key : "))
li = []
st = ""
for i in range(len(s)):
    li.insert(i,s[i])   
for j in range(k):
    c = j
    while c<len(li):
        st += li[c]
        c+=k
print(st)        
```

```python
# Rail fence cipher 
# input : tssya/h__tn/iiph = tssyah__tniiph
# output : "this is python"
s = input("Enter the string : ")
k = int(input("Enter the key : "))
li = []
l = []
i = 0
for i in range(len(s)):
    li.insert(i,s[i])
    l.insert(i,s[i])
print(li,l)    
for j in range(k):
    c = j
    while i<len(l) :
        l[c] = li[i]
        c=c+k
        i+=1
print(li,l)
```

```python
# Rail fence cipher 
# input : tssya/h__tn/iiph = tssyah__tniiph
# output : "this is python"
s = input("Enter the string : ")
k = int(input("Enter the key : "))
part = len(s)//k
extra = len(s)%k
l = [""]*len(s)
count=0
for i in range(k):
    if i <extra:
        l[i::k]=s[count:count+part+1]
        count=count+part+1
    else:
        l[i::k]=s[count:count+part]
        count=count+part
print("".join(l))        
```
