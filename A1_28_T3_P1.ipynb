```python
# WAP to remove all the comments from a given python code in a text file

file = input("Enter the filename : ")
f = open(file,"r",encoding="utf-8")
f2 = open("new_pro1","w",encoding="utf-8")

for i in f:
    res=""
    for char in range(len(i)):
        if i[char][0][0]=="#":
            break
        res+=i[char]    
    print(res," ") 
    r = res+" "+"\n"
    f2.write(r)
f2.close()
f.close()
```
