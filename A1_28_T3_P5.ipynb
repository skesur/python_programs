```python
class ATM:
    def __init__(self):
        self.pin=""
        self.balance=20000
    def menu(self):
        user = input('''
        1. Press 1 to create pin,
        2. Press 2 to change pin,
        3. Press 3 to check balance,
        4. Press 4 to withdraw,
        5. Press anything to exit
        ''')
        if user=="1":
            self.create_pin()
        elif user=="2":
            self.change_pin()
        elif user=="3":
            self.check_balance()
        elif user=="4":
            self.withdraw()
        else:
            print("Thank you")
    def create_pin(self):
        pin = input("Enter 4 digit pin: ")
        self.pin = pin
        print("Pin created successfully")
        self.menu()
    def change_pin(self):
        pin = input("Enter 4 digit pin: ")
        if pin!=self.pin:
            print("Invalid pin")
            self.menu()
        new_pin=input("Enter New Pin: ")
        self.pin=new_pin
        print("Pin changed successfully")
        self.menu()
    def check_balance(self):
        pin = input("Enter 4 digit pin: ")
        if pin!=self.pin:
            print("Invalid pin")
            self.menu()
        print(f"Current pin: {self.balance}")
        self.menu()
    def withdraw(self):
        pin = input("Enter 4 digit pin: ")
        if pin!=self.pin:
            print("Invalid pin")
            self.menu()
        bal = int(input("Enter balance: "))
        if bal<=self.balance:
            self.balance-=bal
            print(f"{bal} withdraw successfully, current balance {self.balance}")
        else:
            print("Insufficient balance")
            self.menu()
```

```python
atm = ATM()
atm.menu()
```
